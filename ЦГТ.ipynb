```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

# Было выбрано експотенциальное распределение

$f_x=\lambda\times e^{-\lambda x}, x>0$   $ F_x=1 - e^{-\lambda x}$ 

$$Mx = 1/ \lambda$$
$$Dx = 1/ {\lambda}^2$$

```python
lambd = 30 # интенсивность
```

```python
Mx = 1 / lambd # мат ожидание
Dx = 1 / (lambd ** 2) #
s = Dx ** (1/2) # среднеквадратическое отклонение
```

```python
expon = stats.expon(scale=1/lambd)
```

```python
# зададим нормально распределенную случайную величину
norm_rv = stats.norm
# 
x_axis = np.arange(-0.2, 0.2, 0.001)
# Функция плотности
pdf = norm_rv.pdf(x_axis, loc=Mx, scale=s)
```

## n=5

```python
n = 5
#вычисляем средние
y = np.array([sum(expon.rvs(n)) / n for i in range(1000)])
fig, ax1 = plt.subplots(1, 1)
ax2 = ax1.twinx()
ax1.hist(y, normed=True, bins=20)
# нормальное распределений, аппроксимирующих выборочные средние при n=5
ax2.plot(x_axis, pdf)
plt.title('n=5')
plt.show()
```
